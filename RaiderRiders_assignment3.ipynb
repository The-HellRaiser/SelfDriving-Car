With this intention in mind, we decided to work on simulating a racing car environment, which is complex as we need to identify lanes, generate reward functions to help keep the car in the same lane or penalize it when it goes off the road, and identify and plan a path to ensure it gets back on track with the least amount of time.
* To achieve our goal, we can use the Gymnasium's Car Racing-v2 environment, which gives us the ability to effectively play the environment by learning continuous control actions from high-dimensional visual observations. Challenges include continuous action space, sparse rewards, complex dynamics, and the exploration-exploitation trade-off. Various reinforcement learning algorithms like Deep Q-Networks (DQNs) and Proximal Policy Optimization (PPO) are explored and evaluated based on the average rewards achieved.

<h3>Assignment 3</h3></span>

* The gym car racing environment involves continuous control of a car's steering and acceleration to navigate a track, with the agent receiving pixel observations and needing to learn a policy that maps observations to continuous actions. Previously, DQN and DDQN solved this by discretizing actions, but being designed for discrete spaces, they can lead to suboptimal performance in continuous control tasks. To solve this we plan to implement Policy Gradient algorithms like DDPG, SAC. while implmenting this we encounter many issues like convergence, stablility, High compute requirement.
* The problem involves several challenges: dealing with high-dimensional pixel observations from the car racing environment, which makes learning directly from raw data difficult for neural networks; balancing the exploration-exploitation trade-off, where the agent needs to explore new states/actions while exploiting its current policy to maximize rewards; the credit assignment problem of learning which actions lead to long-term rewards, challenging in continuous control with delayed rewards; and stability/convergence issues like diverging or oscillating Q-values that can hinder performance of Policy Gradient algorithms and make achieving good results challenging.

It requires only a straightforward actor-critic architecture and learning algorithm
  * Future Research Solution Ideas

<h2>Assignment 2</h2>
</span>

* We employed the Deep Q-Network (DQN) algorithm to train our Convolutional Neural Network (CNN) for the Car Racing environment. During the training process, we observed a gradual increase in the rewards as the number of episodes increased. To enhance the performance of our agent, we incorporated a technique commonly used in Atari games, which involves stacking consecutive frames. This approach proved to be extremely beneficial, as it provides temporal information that can be crucial for tasks involving motion or dynamics.

* Despite the additional computational cost associated with combining frame stacking and DQN, the rewards generated were impressive. After training for merely 100 episodes, we were able to achieve an average reward of 125. Extending the training to 600 episodes yielded an average reward of 675, demonstrating the effectiveness of this approach. 

* It's worth noting that training a DQN was computationally expensive, requiring more than 72 hours of training time in our case. This highlights the demanding nature of such algorithms and the potential need for optimizations or access to more computational resources.

* In an effort to explore alternative approaches, we implemented a Prioritized Double DQN with Experience Replay. This model was coupled with a more complex grayscale image processing technique, which involved operations like cropping and creating masks to detect specific features, such as the visibility of the road. By focusing on extracting relevant image features, this method allowed us to concentrate on the most critical aspects of the environment.

* While the complex grayscale processing technique proved advantageous in terms of feature extraction, it initially yielded low rewards. It wasn't until the agent was trained for over 300 episodes that we started to observe significant rewards. To fully realize the potential of this approach, we extended the training to 900 episodes, ultimately achieving rewards around 850.

* The Figure above shows how the rewards are accumulated with trianing for DQN


* Throughout our experimentation, we observed trade-offs between computational complexity, training time, and the effectiveness of the employed techniques. The frame stacking approach, while computationally expensive, demonstrated impressive results early on in the training process. On the other hand, the complex grayscale processing technique required more extensive training but offered the potential for better feature extraction and ultimately higher rewards.

<h2>Assignment 3</h2>
</span>

* Since the car racing env is a continouous action space it became the DQN and its variants were not the right choice as it was designed for discrete action spaces. We decided to implement policy gradient methods like Deep deterministic policy gradient algorithm(DDPG) and PPO policy proximal Optimisation. DDPG being an off policy method and PPO being an On policy mehtod gave us good uinderstanding as to how the environment would mbehove with different kinds of policies  
* The DDPG algorithm implemented in this code follows the actor-critic framework, where the actor network learns the policy (mapping states to actions), and the critic network learns to evaluate the value of state-action pairs. The use of target networks and experience replay contributes to stable and efficient learning.
* Effect of noise generator is significant in the DDPG algorithms it is crucial for promoting exploration by adding temporally correlated noise to the actions selected by the actor network. It uses the Ornstein–Uhlenbeck process, allows configurable mean and standard deviation, supports decaying noise, and can be reset. Proper noise generation gave us effective exploration-exploitation trade-off during training.



* Fig explains how rewards are acheived over 800 episodes
* We also implmented PPO which is an on policy method. It uses better trust region for fewer policy updates, The exploration takes on slowly but catches up later very quickly.
* The policy network in PPO outputs a probability distribution over the continuous action space, which can be modeled using a Gaussian distribution or a Beta distribution.  We used these distributions to select the most appropriate action and also lead to convergence.



* Fig explains how rewards are acheived over 500 episodes
* The rewards initially increase due to effective exploration. However, they can drop later could be due to many reasons like premature convergence, overfitting , imbalance between exploration-exploitation,which is leading to performance degradation.As we come to 500 episodes of training we get desired rewards. 

* When performnce of both DDPG and PPO are compared with eachother we realise that the inital exploration rate of DDPG was higher and we were able to secure high rewards where as the on policy PPO was similar. But we could see a drop in rewards as the training had progressed in the PPO algorithm, which could be due to catastrophic forgetting or need to further tune the parameters, where as with DDPG the rewards increae had gradually slowed down and with triaing the difference between the Avg and Max reward had converged. PPO also begins to give satisfactory rewards after 500 episodes of training.

* We trained PPO and DDPG for  more than 8 hours, which is similar in trianing time compared to DQN and DDQN. The training time would differe significantly with the hardware and hyper paramter tuning of the models.

```python
import gymnasium as gym

import numpy as np
```

```python
#block name- Ranomm policy

env = gym.make("CarRacing-v2", domain_randomize=True, render_mode="human")

observation = env.reset()
env.metadata['render_fps'] = 200 

done=False
while not done:
    total_reward = 0
    
    # Run the episode
        # Render the environment
    env.render()
    
    # Take a random action
    action = env.action_space.sample()
    #action = np.array([1.0, 0.0, 0.0])
    
    # Step the environment with the random action
    observation, reward, done, info,_ = env.step(action)
    
    # Update the total reward
    total_reward += reward
    
    #print(f"Episode {episode+1} total reward: {total_reward:.2f}")
env.close()




```
